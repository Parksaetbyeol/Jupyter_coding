1. 모델만들기
2. 전처리한 데이터 집어넣고 학습 (전처리를 잘하면 성능도 향상됨)
3. 새로운 데이터 예측하기
4. 성능향상을 위해 어떻게 할지 더 공부
- tf.keras.models.Sequential([]) : 딥러닝 모델 만들기 
- tf.keras.layers.Dense(노드갯수) : hidden layer생성
- model.compile(optimizer=, loss= ,metrics=[]) : 학습방식 환경 설정해주기
    - optimizer='adam' : adam을 가장 많이 쓴다
    - binary_crossentropy:0과 1사이의 분류,확률예측문제에서 가장 많이 사용.외우기!
    - metrics(척도)=accuracy(정확도, 가장많이씀)
    - loss : 최적화 과정에서 최소화될 손실함수를 설정하는것 (mse,binary_crossentropy자주사용)
------------------------------------------------------------------------------------
- model.fit(x,y,batch_size=, epochs=) : 학습시키기
    - x=학습데이터, y=결과데이터
    - batch_size=한번에 학습할때 사용하는 데이터 갯수
    - epochs=학습 데이터 반복 횟수

### dense에 들어가는 파라미터
- units : 출력 값의 크기
- activation : 활성화 함수
- use_bias : 편향(b)을 사용할지 여부
- kernel_initializer : 가중치(W) 초기화 함수
- bias_iniotializer : 편향 초기화 함수
- kernel_regularizer : 가중치 정규화 방법
- bias_regularizer : 편향 정규화 방법
- activity_regularizer : 출력 값 정규화 방법
- kernel_constraint : 가중치에 적용되는 부가적인 제약 함수
- bias_constraint : 편향에 적용되는 부가적인 제약 함수

## Q1. 키와 신발사이즈는 어떤 관련이 있을까?
- 키 = [170,180,175,160]
- 신발 = [260,270,265,255]

```python
# tf.Variable(초기값) : w값(변수) 만듬
a = tf.Variable(0.1)
b = tf.Variable(0.2) # a,b좋은결과 나올떄까지 학습할거임
```

```python
def loss_function():
    예측값 = 키 * a + b
    return tf.square(260- 예측값)
```

```python
# tf.keras.optimizers : 경사하강법을 이용해서 변수들을 업데이트해주는 함수
# Adam : 각 파라미터마다 다른 크기의 업데이트를 적용
opt=tf.keras.optimizers.Adam(learning_rate=0.1)
```

```python
for i in range(300): # 300번 반복해줘
    opt.minimize(loss_function, var_list=[a,b]) # var_list : 경사하강법으로 업데이트할 weight variable 목록
    print(a.numpy(),b.numpy()) # .numpy(): numpy값만 출력 -> a:1.519 b:1.619
```

```python
# 찾은 a:1.52, b:1.62 값을  '신발 = 키 * a + b' 식에 대입해보기!
170 * 1.52 + 1.62
# 키 170일때 신발사이즈가 260이라고 예측이 가능해졌다!
```
