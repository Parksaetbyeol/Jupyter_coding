# 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'  # 폰트지정
mpl.rcParams['font.size'] = 15 # 글자크기 지정
mpl.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스(-) 글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

## subplots
좌표를 한꺼번에 표현

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10)) # 2x2(총 4개)에 해당하는 plot들을 생성 (2,3)하면 2x3형태의 6개의 plot들 생성
fig.suptitle('여러 그래프 넣기')

# 첫번째 그래프 지정
axs[0,0].bar(df['이름'], df['국어'], label='국어점수') # 데이터 설정
axs[0,0].set_title('첫 번째 그래프') # 제목
axs[0,0].legend() # 범례
axs[0,0].set(xlabel='이름', ylabel='점수') # x,y축 label
axs[0,0].set_facecolor('lightyellow') # 전면 색
axs[0,0].grid(linestyle='--', linewidth=0.5)

# 두번째 그래프 지정
axs[0,1].plot(df['이름'],df['수학'], label='수학')
axs[0,1].plot(df['이름'],df['영어'], label='영어')
axs[0,1].legend()

# 세번째 그래프 지정
axs[1,0].barh(df['이름'],df['키'])

# 네번째 그래프 지정
axs[1,1].plot(df['이름'],df['사회'], color='g', alpha=0.5, label='사회')
axs[1,1].legend(loc='lower right')
```

## subplot
좌표를 하나씩 표현

```python
plt.subplot(2,1,1) # 2행 1열의 1번째 그래프
plt.subplot(2,1,2) # 2행 1열의 2번째 그래프
plt.show()
```

```python
import numpy as np
x = np.arange(1,10)
fig = plt.figure()
ax1 = plt.subplot(2,1,1)
ax2 = plt.subplot(2,1,2)

ax1.plot(x)
ax2.plot(x,color='r')
plt.show()
```
