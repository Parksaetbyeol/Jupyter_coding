# 8. 막대 그래프(기본)
### - plt.bar(x축의 데이터, y축의 데이터)

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'  # 폰트지정
mpl.rcParams['font.size'] = 15 # 글자크기 지정
mpl.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스(-) 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호','서태호','정대만']
values = [190, 187, 184] # 키
plt.bar(labels,values, color='r') # plt.bar(x축의 데이터, y축의 데이터)
```

```python
labels = ['강백호','서태호','정대만']
values = [190, 187, 184] # 키
colors= ['r','g','b']
plt.bar(labels,values, color=colors, alpha=0.5)
```

```python
labels = ['강백호','서태호','정대만']
values = [190, 187, 184] # 키
plt.bar(labels,values)
plt.ylim(175,195) # y축의 범위를 제한
```

```python
plt.bar(labels,values, width=0.5)
```

```python
plt.bar(labels,values, width=0.3)
plt.xticks(rotation=45) # x축의 이름데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y축의 이름데이터 각도를 45도로 설정
plt.show()
```

```python
labels = ['강백호','서태호','정대만']
values = [190, 187, 184]
ticks = ['1번학생','2번학생','3번학생']
plt.bar(labels,values)
plt.xticks(labels,ticks, rotation=90)
```
