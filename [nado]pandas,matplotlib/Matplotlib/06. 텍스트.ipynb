# 텍스트

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'  # 폰트지정
mpl.rcParams['font.size'] = 15 # 글자크기 지정
mpl.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스(-) 글자가 깨지는 현상을 해결
```

```python
x= [1,2,3] # x축 리스트 데이터
y= [2,4,8] # y축 리스트 데이터
```

```python
plt.plot(x,y)
```

### enumerate
- 반복문 사용 시 몇 번째 반복문인지 확인이 필요할 수 있습니다. 이때 사용합니다.
- 인덱스 번호와 컬렉션의 원소를 tuple형태로 반환합니다.

```python
t = [1, 5, 7, 33, 39, 52]
for p in enumerate(t): # t데이터를 p로 받아서
    print(p) # 인덱스번호와 p를 출력
```

```python
for i, v in enumerate(t):
    print("index : {}, value: {}".format(i,v))
```

### 텍스트 삽입 함수 plt.text()
- plt.text(x, y, s,fontdict=None, **kwargs) 
    - x와y는 텍스트를 배치해야하는 좌표 / s는 추가해야하는 텍스트의 내용/ fontdict 텍스트의 속성을 설정 
    

```python
plt.plot(x, y, marker='o')
for idx, txt in enumerate(y): # y의 2,4 8을 하나씩 txt에 넣고, inx는 0,1,2.. 순서를 보여줌
    plt.text(x[idx], y[idx], txt) # x의 0번째인덱스(1), 1번째인덱스(2), 2번째 인덱스(3)가져오고, y도 똑같이 가져옴, y를 받은 txt를 입력
```

```python
plt.plot(x, y, marker='o')
for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx] + 0.3, txt, ha='center') # y에 0.3더해서 txt 좀 위로 올리기 # ha 수평맞춰줌
```

```python
plt.text(0.55, 0.55, 'Hello World!', fontsize=20, color='green') # x,y축이 최대 1.0이니까 그 이상의 위치에 텍스트를 입력할 수 없다.
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.text(4,6, 'Hello World!', fontsize=20, color='green')
plt.xlim([0,10]) # x축,y축의 제한범위를 지정해서 더 넓혀준다.
plt.ylim([0,10])

plt.show()
```

### kwargs 이용해서 텍스트에 다양한 스타일 추가
1) verticalalignment

  - center : 수직방향으로 좌표 중앙에 놓임

  - top : 좌표가 텍스트 위에 놓임

  - bottom : 좌표가 텍스트 아래 놓임

  - baseline : 텍스트의 baseline 에 따라 달라짐


2) horizontal alignment

  - center : 수평방향으로 좌표 중앙에 놓임

  - left : 좌표가 텍스트 왼쪽에 놓임

  - right : 좌표가 텍스트 오른쪽에 놓임 

```python
plt.text(5,5,"Hello World!",
         fontsize=20,
         color="red",
         verticalalignment ='top', 
         horizontalalignment ='center',
         bbox ={'facecolor':'grey', 
                'pad':10}
        )                
         
plt.xlim([0,10])
plt.ylim([0,10])
plt.show()
```
