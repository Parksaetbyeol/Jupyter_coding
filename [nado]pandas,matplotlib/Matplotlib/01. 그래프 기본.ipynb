# 1. 그래프 기본
Basic Line Plot 선그래프

```python
x= [1,2,3] # x축 리스트 데이터
y= [2,4,8] # y축 리스트 데이터
plt.plot(x,y)
```

```python
print(plt.plot(x,y))
```

```python
plt.plot(x,y)
plt.show() # 그래프위에 텍스트 안보임
```

## 한글 폰트 설정
- import matplotlib as mpl
- mpl.rcParams['font.family'] = '원하는 폰트 적기'

```python
# 한글폰트 사용시 필수
import matplotlib as mpl
mpl.rcParams['font.family'] = 'HYGungSo-Bold'  # 폰트지정
mpl.rcParams['font.size'] = 15 # 글자크기 지정
mpl.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스(-) 글자가 깨지는 현상을 해결
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] # 폰트 리스트에서 폰트 이름만 출력
```
