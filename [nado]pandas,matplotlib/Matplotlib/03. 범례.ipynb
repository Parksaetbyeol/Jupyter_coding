# 3. 범례(legend)
label='범례'

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'  # 폰트지정
mpl.rcParams['font.size'] = 15 # 글자크기 지정
mpl.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스(-) 글자가 깨지는 현상을 해결
```

```python
x= [1,2,3] # x축 리스트 데이터
y= [2,4,8] # y축 리스트 데이터
```

### 1.  label로 해당 그래프의 범례텍스트를 미리 지정한뒤 legend로 실행

```python
plt.plot(x, y, label='무슨 데이터') # label='범례'
plt.xlabel('x축')
plt.ylabel('y축')
plt.legend()
```

```python
plt.plot(x, y, label='무슨 데이터') # label='범례'
plt.legend(loc='lower right') # 범례 위치 # 위치코드 lower right=4
```

```python
plt.plot(x, y, label='무슨 데이터') # label='범례'
plt.legend(loc='upper left') # 위치 : 'upper right' 'lower left' 'center left' 'center' 등
```

```python
plt.plot(x, y, label='무슨 데이터') # label='범례'
plt.legend(loc='center left')
```

```python
plt.plot(x,y,label='범례')
plt.legend(loc=(0.5, 0.5)) # loc=(x축기준,y축기준) (0~1사이)
```

```python
import numpy as np
x = np.linspace(-2 * np.pi, 2*np.pi, 100) # np.linspace(start,stop,num)
# label로 범례 텍스트 지정
plt.plot(x, np.sin(x), color='r', label = 'y=sin x')
plt.plot(x, np.cos(x), color='b', label = 'y=cos x')
plt.plot(x, x, color='g', label='y=x')
plt.legend(loc='upper left', fontsize=12)
plt.show()
```

### 2. legend메소드 내에서 그래프 순서대로 텍스트 지정

```python
# 위 그래프와 같은 결과 출력
plt.plot(x, np.sin(x), color = 'red')
plt.plot(x, np.cos(x), color = 'blue')
plt.plot(x, x, color = 'green')

# 범례 텍스트를 따로 지정
plt.legend(['y = sin x', 'y = cos x', 'y = x'], loc='lower right')
plt.show()
```
