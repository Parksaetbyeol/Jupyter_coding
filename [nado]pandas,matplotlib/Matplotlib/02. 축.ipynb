# 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'  # 폰트지정
mpl.rcParams['font.size'] = 15 # 글자크기 지정
mpl.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스(-) 글자가 깨지는 현상을 해결
```

```python
x= [1,2,3] # x축 리스트 데이터
y= [2,4,8] # y축 리스트 데이터
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size' : 20}) # fontdict={} 개별 폰트 설정
```

```python
plt.plot(x,y)
plt.xlabel('x축') # x축 이름 설정
plt.ylabel('y축') # y축 이름 설정
```

```python
plt.plot(x,y)
plt.xlabel('x축', color='red') # 색 추가
plt.ylabel('y축', color='#00aa00') # '#00aa00'은 RGB중에서 R은(00)없고, G는 aa값, B는(00) 없음
```

```python
plt.plot(x,y)
plt.xlabel('x축', color='red', loc='right') # x축이름은 right,left,center 위치 세가지중 선택
plt.ylabel('y축', color='#00aa00',loc='top') # y축 이름은 top, center, bottom 위치 세가지중 선택
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])  # x축에 표시되는 눈금 값 지정
plt.yticks([3,6,9,12]) # y축에 표시되는 눈금 값 지정
plt.show()
```

```python
plt.plot(x,y)
xtick=[1,2,4]
ytick=[3,6,9,10]
plt.xticks(xtick)
plt.yticks(ytick)
plt.show()
```
