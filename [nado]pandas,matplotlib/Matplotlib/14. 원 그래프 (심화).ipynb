# 14. 원 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'  # 폰트지정
mpl.rcParams['font.size'] = 15 # 글자크기 지정
mpl.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스(-) 글자가 깨지는 현상을 해결
```

## 퍼센트 나타내기
autopct='%.1f%%' 소수점 첫째자리까지 퍼센트 나타내기

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python','Java','Javascript','C#','C/C++','ETC']
#colors = ['b', 'g', 'r', 'c', 'm', 'y']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode= [0.05]*6

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, explode=explode) # counterclock 12시 기점으로 시계방향으로 할지 반시계방향으로 할지
plt.show()
```

## 부채꼴 스타일 wedgeprops

```python
wedgeprops={'width':0.5} # 1에 가까울수록 원 가운데에 가까워지고, 0에 가까울수록 테두리에 가꺼워짐
plt.pie(values,labels=labels,autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, explode=explode, wedgeprops=wedgeprops)
plt.show()
```

```python
# 위에 explode한게 이상해서 삭제함
wedgeprops={'width':0.5}
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,counterclock=False,colors=colors,wedgeprops=wedgeprops)
plt.show()
```

```python
# 테두리색상을 하얀색으로 해서 그래프들끼리 떨어진것처럼 보이게함
wedgeprops={'width':0.5, 'edgecolor':'w', 'linewidth':5}
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,counterclock=False,colors=colors,wedgeprops=wedgeprops)
plt.show()
```

### 소수점자리 지정해서 값 입력 
- 수식) '%.1f' % 나타내려는 값 : % <-꼭쓰고 뒤에 나타내려는 값 입력

```python
# 함수로 만들어서 10 이상이면 소수점 첫번째까지 퍼센트 %표시
def custom_autopct(pct): # pct데이터를 받아서
    return('%.1f%%' % pct) if pct >= 10 else '' # pct가 10 이상이면 '%.1f%%'(소수점첫째자리까지 %표시)를 반환, 아니라면 빈칸 반환
plt.pie(values,labels=labels,autopct=custom_autopct,startangle=90,counterclock=False,colors=colors,wedgeprops=wedgeprops)
plt.show()
```

```python
def custom_autopct(pct): # pct데이터를 받아서
    return'{:.1f}%'.format(pct) if pct >= 10 else '' # 위와 똑같이 다르게도 쓸수있음
plt.pie(values,labels=labels,autopct=custom_autopct,startangle=90,counterclock=False,colors=colors,wedgeprops=wedgeprops)
plt.show()
```

## 퍼센트 표시 위치 변경
pctdistance= 0~1 (0에 가까울수록 가운데로 모인다. 1에 가까울수록 밖으로)

```python
def custom_autopct(pct): # pct데이터를 받아서
    return'{:.0f}%'.format(pct) if pct >= 10 else '' # 퍼센트가 10이상이면 그대로 출력, 아니면(10미만) 빈칸출력
plt.pie(values,labels=labels,autopct=custom_autopct,startangle=90,counterclock=False,colors=colors,
        wedgeprops=wedgeprops, pctdistance=0.7)
plt.show()
```
