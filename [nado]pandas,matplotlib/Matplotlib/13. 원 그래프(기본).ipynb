# 13. 원 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'  # 폰트지정
mpl.rcParams['font.size'] = 15 # 글자크기 지정
mpl.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스(-) 글자가 깨지는 현상을 해결
```

```python
#values = [30, 25, 20, 13, 10, 2]
values = [1,1,1,1,1,1] # 꼭 합이 100이 아니어도 자동으로 100%기준으로 나눠준다
labels = ['Python','Java','Javascript','C#','C/C++','ETC']

plt.pie(values, labels=labels, autopct='%.1f') # autopct 퍼센트 표시
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python','Java','Javascript','C#','C/C++','ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90) # %%하면 숫자뒤에 '%'표시해줌 # startangle 시작각도를 정해줌
plt.show()                                                                                    # (90도부터 반시계방향으로)
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python','Java','Javascript','C#','C/C++','ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) # counterclock=False 시계방향으로 그래프 그려짐
plt.show()
```

```python
# explode 간격띄우기
values = [30, 25, 20, 13, 10, 2]
labels = ['Python','Java','Javascript','C#','C/C++','ETC']
#explode = [0.2, 0.1, 0, 0, 0, 0]
explode = [0.05] * 6
plt.pie(values,labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2,0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2,0.3), title='언어별 선호도')
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.title('언어별 선호도')
plt.legend(loc=(1.2,0.3), title='언어별 선호도')
plt.show()
```
