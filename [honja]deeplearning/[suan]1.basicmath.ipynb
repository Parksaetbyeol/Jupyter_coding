## ÏùºÏ∞®Ìï®Ïàò
- y = aùí≥+b
    - a : Í∏∞Ïö∏Í∏∞, b: yÏ†àÌé∏
- Í∑∏ÎûòÌîÑ ÏÉÅÏóêÏÑú ÏßÅÏÑ†Ïù∏ Í∑∏ÎûòÌîÑ(linear)

```python
# linear Ìï®Ïàò ÎßåÎì§Í∏∞
def linear_function(x):
    a = 0.5
    b = 2
    return a*x + b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5,5,0.1) # arange(start,stop,step)
y = linear_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.show()
```

## Ïù¥Ï∞®Ìï®Ïàò
- y = aùí≥¬≤+bùí≥ + c
- ÏùºÎ∞òÏ†ÅÏúºÎ°ú ÎëêÍ∞úÏùò Ïã§Í∑ºÏùÑ Í∞ÄÏßê

```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    return a*x**2 + b*x + c
```

```python
print(quadratic_function(2))
```

```python
x = np.arange(-5,5,0.1)
y = quadratic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
plt.show()
```

## ÏÇºÏ∞®Ìï®Ïàò
- y = aùí≥¬≥+bùí≥¬≤ + cùí≥ + d

```python
def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8
    return a*x**3 + b*x**2 + c*x + d
```

```python
print(cubic_function(3)) # 4*27 + 0 + (-3) + (-8)
```

```python
x = np.arange(-5,5,0.1)
y = cubic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
plt.show()
```

### ÎØ∏Î∂ÑÍ≥º Í∏∞Ïö∏Í∏∞
- ÎØ∏Î∂ÑÏùÄ 'ÌäπÏ†ï ÏàúÍ∞Ñ'Ïùò Î≥ÄÌôîÎüâÏùÑ ÎúªÌïúÎã§.

```python
def numerical_diff(f,x):
    h = 1e-4
    return (f(x+h)-f(x-h)/2*h)
```

```python
h = 1e-4
h
```

```python
def my_func(x):
    return 2*x**2 + 3*x + 2
```

```python
def linear_func(a,b,c,x): # a=xÏ¢åÌëú, b=yÏ¢åÌëú, c=Í∏∞Ïö∏Í∏∞, x=Í∞í
    return c*(x-a) + b
```

```python
c = numerical_diff(my_func,1)

x=np.arange(-5,5,0.01)
y=linear_func(1,my_func(1),c,x)

plt.xlabel('x')
plt.ylabel('y')
plt.scatter(1,my_func(1)) # Ï†ëÏ†êÏóê Ï†êÏ∞çÍ∏∞
plt.plot(x,my_func(x),x,y,'r-')
plt.title('f(x) & linear function')
plt.show()
```

### 'Í∏∞Ïö∏Í∏∞' Íµ¨ÌïòÎäî Ìï®Ïàò Íµ¨ÌòÑÌïòÍ∏∞

```python
def gradient(f,x):
    h=1e-5
    grad = np.zeros_like(x)
    
    for i in range(x.size): # x_sizeÎßåÌÅº Î∞òÎ≥µ =2
        tmp = x[i]
        x[i] = tmp + h
        fxh1 = f(x) # fxh1Ïóê fÌï®Ïàò Í≤∞Í≥ºÍ∞í ÎÑ£Í∏∞
        
        x[i] = tmp - h
        fxh2 = f(x)
        
        grad[i] = (fxh1 - fxh2) / (2*h)
        x[i] = tmp
        
    return grad
```

```python
print(gradient(my_func2, np.array([3.0,4.0])))
print(gradient(my_func2, np.array([1.0,2.0])))
```
