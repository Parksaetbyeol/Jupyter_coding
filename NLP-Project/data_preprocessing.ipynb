### 커스텀 함수

```python
# 데이터프레임을 텍스트 파일로 저장하는 함수
def DFtofile(DFtofile,resname):
    a=pd.DataFrame(DFtofile)
    a.to_csv('_res/'+resname+'.txt')
```

```python
# 피벗테이블을 통해 중복 Y값이 있는 카테고리를 탐색
def pivot_check(df,names):
    temp=df[['crdCoBzkNm','sndCtgyNm']].groupby(['crdCoBzkNm', 'sndCtgyNm'])[['sndCtgyNm']].count()
    temp.columns=['count']
    temp.reset_index(inplace=True)
    #temp.pivot(index='crdCoBzkNm', columns='sndCtgyNm', values='count').fillna(0)                 # 방법 1
    temp=temp.pivot_table(index='crdCoBzkNm', columns='sndCtgyNm', values='count', aggfunc='sum')  # 방법 2
    temp['res']=temp.apply('count',axis=1)
    temp=temp.sort_values(by='res',ascending=False)
    temp.to_excel('_res/'+'(python)pivot_check_'+names+'.xlsx')
    return temp[temp['res']>1]    
```

```python
# 레벨링
def EncodingLabel(x):
    le = LabelEncoder()
    le.fit(x)
    label = le.transform(x)
    return label
```

```python
# 토큰화
def getPPS(x):
    res=[]
    pps=['NNG','NNP','NNB','NNBC','NR','NP','VV','VA','VX',
         'VCP','VCN','MM','MAG','MAJ','IC','JKS','JKC','JKG',
         'JKO','JKB','JKV','JKQ','JX','JC','EP','EF','EC',
         'ETN','ETM','XPN','XSN','XSV','XSA','XR','SF',
         'SE','SSO','SSC','SC','SY','SL','SH','SN']  # 모든 품사 지정
    pos=mecab.pos(x)
    #print(pos)
    for po in pos:
        if (po[1] in pps):
            res.append(po[0])
    return res
```
