```python
# 35번 라벨을 21번으로 변경
train_y = train_y.replace(35, 21)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(learning_rate=0.1, max_depth=3)
model.fit(train_X, train_y)
```
