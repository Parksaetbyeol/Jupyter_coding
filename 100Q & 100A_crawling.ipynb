# 100문 100답
- https://www.easylaw.go.kr/CSP/OnhunqueansLstRetrieve.laf?onhunqnaAstSeq=84&search_put=

```python
import requests 
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
page = requests.get("https://www.easylaw.go.kr/CSP/OnhunqueansLstRetrieve.laf?onhunqnaAstSeq=84&search_put=")
print(page.status_code) # 200
print(page.text)
```

```python
question = []
answer = []

for page_num in range(13):
    url = f'https://www.easylaw.go.kr/CSP/OnhunqueansLstRetrieve.laf?curPage={page_num+1}&search_put=&astSeq=QNA84&onhunqnaAstSeq=84&onhunqueAstSeq=84&pageType=20'
    
    response = requests.get(url)
    soup = bs(response.text , 'html.parser')

    html_que = soup.select('div.ttl a')
    html_ans = soup.select('div.ans p')

    
    for i in html_que:
        question.append(i.get_text())
        
    for i in html_ans:
        answer.append(i.get_text())

list_sum = list(zip(question, answer))


col = ['질문','답변']

df = pd.DataFrame(list_sum, columns=col)
```

```python
df.질문 = df.질문.str.replace('\r\n','').str.strip()
```

# 한국 부동산원 자주 묻는 질문

```python
page = requests.get('https://www.reb.or.kr/reb/na/ntt/selectNttList.do?mi=9665&bbsId=1152&srch_aditCol1=%EB%B6%80%EB%8F%99%EC%82%B0%EA%B3%B5%EC%8B%9C')
print(page.status_code) # 200
print(page.text)
```

```python
que_list = []

soup = bs(page.text, "html.parser")

elements = soup.select('div.BD_list a')
        
for i in elements:
    que_list.append(i.get_text())

list_sum = list(que_list)


col = ['질문']

df2 = pd.DataFrame(list_sum, columns=col)
```

```python
df2.질문 = df2.질문.str.replace('\r\n\t\t\t\t\t\t\t\t\t\t\t','').str.strip()
df2
```

```python
df2.to_excel('한국부동산원 자주 묻는 질문.xlsx')
```
